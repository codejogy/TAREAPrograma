```python
# Constantes
ao = 340.29 #m/s
po = 101325 #pa
IAS = 280 #m/s
CAS = 277 #m/s
#qc = 15347 #pa

# Press ratios
pratios = [0.8637,
0.8014,
0.7428,
0.6877,#10000 ft
0.6360,
0.5856,
0.5420,
0.4994,
0.4595,#20000 ft
0.4223,
0.3876,
0.3552,
0.3250,
0.2970] # 3000ft

# Density ratio

dratios = [0.88809,
0.83585,
0.78602,
0.73848,
0.69317,
0.65003,
0.60896,
0.56991,
0.53281,
0.49758,
0.46417,
0.43250,
0.40251,
0.37413]

i = 4000 
alturas = []
for step in range(4000,30001,2000):
    alturas.append(step)
print(alturas)

tablaMach = []
tablaEAS = []
tablaTAS = []
```

```python
def Mach(p):
    m = (5*((qc/p+1)**(2/7)-1))
    return m
```

```python
for pressure in pratios:
    p = pressure*po
    mach = Mach(p)
    tablaMach.append(mach)
tablaMach
```

```python
def EAScompr(mach,pratio):
    EAS = ao*mach*(pratio)**(1/2)
    return EAS
```

```python
machpressures = list(zip(tablaMach,pratios))
machpressures
```

```python
for machpressure in machpressures:
    m = machpressure[0]
    prat = machpressure[1]
    EAS = EAScompr(m,prat)
    tablaEAS.append(EAS)
tablaEAS #m/s
    
```

```python
for EAS in tablaEAS:
    print(EAS*1.944) #knots
```

```python
def TAS(EAS,densratio):
    TAS = EAS/((densratio)**(1/2))
    return TAS
```

```python
for easdens in list(zip(tablaEAS,dratios)):
    tas = TAS(easdens[0],easdens[1])*1.944 #knots
    tablaTAS.append(tas)
tablaTAS
```
